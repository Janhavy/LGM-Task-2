```python
# Importing the Libraries
import pandas as pd
import numpy as np
import Recommenders as Recommenders
```

```python
# Loading the data
song_df_1= pd.read_csv("triplets_file.csv")
song_df_1.head()
```

```python
len(song_df_1)
```

```python
song_df_2= pd.read_csv("song_data.csv")
song_df_2.head()
```

```python
len(song_df_2)
```

```python
# Combining the two dataframes
song_df=pd.merge(song_df_1,song_df_2.drop_duplicates(["song_id"]), on = "song_id", how="left")
song_df.head()
```

```python
# Checking the size of dataframe
len(song_df)
```

```python
# Creating the new colum 'song' by combining 'song title' and 'artist name'
song_df["song"]= (song_df["title"]) + " - "  + (song_df["artist_name"])
song_df
```

```python
# Droping unneccessary columns 
song_df = song_df.drop(["title","artist_name"],axis=1)
```

```python
# Taking top 100000 samples in consideration
song_df=song_df.head(100000)
song_df
```

```python
# Aggregating the count of songs
song_grouped=song_df.groupby(["song"]).agg({"listen_count":"count"}).reset_index()
```

```python
song_grouped
```

```python
grouped_sum = song_grouped['listen_count'].sum()
song_grouped['percentage'] = (song_grouped['listen_count'] / grouped_sum ) * 100
song_grouped.sort_values(['listen_count', 'song'], ascending=[0,1])
```

```python
ir = Recommenders.item_similarity_recommender_py()
ir.create(song_df, 'user_id', 'song')
```

```python
user_items = ir.get_user_items(song_df['user_id'][50])
```
